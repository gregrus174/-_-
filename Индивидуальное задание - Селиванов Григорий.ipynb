## 1.1 Удаление "лишних" столбцов

Исходя из данных, некоторые параметры, которые не так важны для рядового пользователя, могут быть:

* Количество ядер процессора (__num_cores__) - аналогично, это более технический параметр, которым рядовой пользователь, как правило, не интересуется.
* Количество потоков, поддерживаемых процессором (__num_threads__) - также технический параметр, не очень важный для большинства пользователей.
* Тип дополнительного накопителя (__secondary_storage_type__) - если он есть, то важно, но если его нет, то рядовой пользователь вряд ли будет сильно об этом беспокоиться.
* Емкость дополнительного накопителя (__secondary_storage_capacity__) - для большинства пользователей этот параметр менее важен, чем емкость основного накопителя.
* Тип видеокарты (__gpu_type__) - рядовых пользователей в большей степени интересует общая производительность, а не конкретный тип видеокарты.

Иными словами, более технические характеристики, которые важны для продвинутых пользователей, но не влияют сильно на повседневное использование ноутбука, могут быть не столь важны для обычного пользователя.

**Итого, я посчитал нужным удалить 5 столбцов за ненужностью содержащейся в них информации.**

```python
df.drop(labels = 'num_cores', axis = 1, inplace = True)
df.drop(labels = 'num_threads', axis = 1, inplace = True)
df.drop(labels = 'secondary_storage_type', axis = 1, inplace = True)
df.drop(labels = 'secondary_storage_capacity', axis = 1, inplace = True)
df.drop(labels = 'gpu_type', axis = 1, inplace = True)
```
